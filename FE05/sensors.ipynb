```python
from elasticsearch import Elasticsearch
import requests as rq
import time

sensors = [
    'http://nosql.hpeixoto.me/api/sensor_elastic/3001',
    'http://nosql.hpeixoto.me/api/sensor_elastic/3002',
    'http://nosql.hpeixoto.me/api/sensor_elastic/3003',
    'http://nosql.hpeixoto.me/api/sensor_elastic/3004',
    'http://nosql.hpeixoto.me/api/sensor_elastic/3005',
    'http://nosql.hpeixoto.me/api/sensor_elastic/4001',
    'http://nosql.hpeixoto.me/api/sensor_elastic/4002',
    'http://nosql.hpeixoto.me/api/sensor_elastic/4003',
    'http://nosql.hpeixoto.me/api/sensor_elastic/4004',
    'http://nosql.hpeixoto.me/api/sensor_elastic/4005'
]

es = Elasticsearch([{'host': 'localhost', 'port': 9200}])

# Insert JSON documents into the 'sensors' index
def insert(data):
    es.index(index='sensors',doc_type='sensor',body=data)

while True:
    print("\nNew Round of GET Requests...")
    for sensor in sensors:
        data = rq.get(sensor).json()
        print("> Currently on sensor: " + sensor.split(sep="/")[5])
        insert(data)
        print("Done!\n")
    time.sleep(5)
```
